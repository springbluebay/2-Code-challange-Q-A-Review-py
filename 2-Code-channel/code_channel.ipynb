```python
# Quick loop & comprehension algorithm practice :
# Given the nested collection below. Convert these collections to the desired (set) output format in a single code line using nested loops.
# Input → x = [([1], [2, 3], (4, 5, 6))]
# Desired out → {1, 2, 3, 4, 5, 6}

x= [([1], [2, 3], [4, 5, 6])]

y=list(x[0])

t=[]

for i in range(len(y)):
    for j in range(len(y[i])):
        t.append(y[i][j])
t=set(t)
print(t)
```
